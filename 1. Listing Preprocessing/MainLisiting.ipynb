```python
def replacer(A):
    import pandas as pd
    Q = pd.DataFrame(A.isna().sum(),columns=["CT"])
    missingcolumns = list(Q[Q.CT > 0].index)
    for i in missingcolumns:
        if(A[i].dtypes=="object"):
            replacer = A[i].mode()[0]
            A[i] = A[i].fillna(replacer)
        else:
            replacer = A[i].mean()
            A[i] = A[i].fillna(replacer)
```

```python
replacer(df)
```
